```python
#lets encode Airline column using Target guided Ordinal encoding technique considering Price as Target feature
df.groupby("Airline")["Price"].mean().sort_values(ascending=False).reset_index()
```

```python
# replace_target_guided_ordinal_encoding
def replace_airline_with_mean():
    mean_prices = df.groupby("Airline")["Price"].mean().sort_values()
    df["Airline"] = df["Airline"].apply(lambda x : mean_prices[x])
    return df
df =replace_airline_with_mean(df)
df.head()
```
